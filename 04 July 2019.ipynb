```python
#Graph to plot distance and time 

distance=list(range(2,11,2))
time=list(range(1,6))
#distance=2*time
plt.title('distance vs time')
plt.plot(time,distance,'bo') # b means blue and o means dot we can also change that colour

```

```python
#Graph to plot distance and time 

distance=list(range(2,11,2))
time=list(range(1,6))
#distance=2*time
plt.title('distance vs time')
plt.plot(time,distance)

```

```python
#Graph to plot distance and time 

distance=list(range(2,11,2))
time=list(range(1,6))
#distance=2*time
plt.title('distance vs time')
plt.xlabel('time')
plt.ylabel('distance')
plt.plot(time,distance,'bo')
plt.grid()
plt.show()
```

###  velocity

```python
# v=at
t=list( range(1,11))
a=list(range(1,len(t)+1))
v=[a[i] * i for i in range (len(t))]

plt.plot(t,v)
plt.grid()
plt.show()
```

### Exponential Increase

```python
n=list(range(10))
sqn=[i**100 for i in n]
plt.plot(n,sqn)
plt.show()
```

```python
n=list(range(10))
sqn=[i**3 for i in n]
plt.plot(n,sqn)
plt.show()
```

```python
#velocity-time
import numpy as np
#t=[i/10 for i in range(1,21)]
t=np.arange(0.1,2.0,0.1)
v=[1+np.sin(2*np.pi*i) for i in t]
plt.plot(t,v)
plt.grid()
plt.xlabel('time')
plt.ylabel('voltage')
plt.show()
```

```python
#velocity-time
import numpy as np
#t=[i/10 for i in range(1,21)]
t=np.arange(0.1,2.0,0.1)
v=[1+np.tan(2*np.pi*i) for i in t]
plt.plot(t,v)
plt.grid()
plt.xlabel('time')
plt.ylabel('voltage')
plt.savefig('images/voltage-time.png')
plt.show()
```
